```python
from sklearn.preprocessing import LabelEncoder
enc=LabelEncoder()
full_train['HomePlanet']=pd.DataFrame(enc.fit_transform(full_train['HomePlanet'].to_numpy()))
full_test['HomePlanet']=pd.DataFrame(enc.transform(full_test['HomePlanet'].to_numpy()))
```

```python
full_train['Destination']=pd.DataFrame(enc.fit_transform(full_train['Destination'].to_numpy()))
full_test['Destination']=pd.DataFrame(enc.transform(full_test['Destination'].to_numpy()))
```

```python
full_train.head()
```

```python
from sklearn.model_selection import train_test_split
Y=full_train['Transported']
X=full_train.drop(['Transported','PassengerId'],axis=1)
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

# XGboost

```python
from sklearn.model_selection import GridSearchCV
```

```python
from xgboost import XGBClassifier
model = XGBClassifier(learning_rate =0.1,
              n_estimators=10,
              max_depth=5,
              min_child_weight=2,
              gamma=0.1,
              subsample=0.85,
              colsample_bytree=0.8,
              objective= 'binary:logistic',
              nthread=4,
              scale_pos_weight=1,
              seed=27)
             
gsearch = GridSearchCV(estimator = model,param_grid= {
 'max_depth':range(3,10,2),
 'min_child_weight':range(1,6,2)}, scoring='roc_auc',n_jobs=4, cv=5,verbose=1
    )


gsearch.fit(x_train,y_train)
print('Best Grid Search Parameters :',gsearch.best_params_)
print('Best Grid Search Score : ',gsearch.best_score_)
```

```python
model2= XGBClassifier(learning_rate =0.1,
              n_estimators=10,
              max_depth=7,
              min_child_weight=5,
              gamma=0,
              subsample=0.85,
              colsample_bytree=0.8,
              objective= 'binary:logistic',
              nthread=4,
              scale_pos_weight=1,
              seed=27)
gsearch2 = GridSearchCV(estimator = model2,param_grid= {
 'gamma':[i/10 for i in range(0,5)]}, scoring='roc_auc',n_jobs=4, cv=5,verbose=5
    )
gsearch2.fit(x_train,y_train)
print('Best Grid Search Parameters :',gsearch2.best_params_)
print('Best Grid Search Score : ',gsearch2.best_score_)
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
model3=XGBClassifier(learning_rate =0.1,
              n_estimators=10,
              max_depth=7,
              min_child_weight=5,
              gamma=0.4,
              subsample=0.85,
              colsample_bytree=0.8,
              objective= 'binary:logistic',
              nthread=4,
              scale_pos_weight=1,
              seed=27)
model3.fit(x_train,y_train)
predict_pr=model3.predict_proba(x_test)
auc_score = roc_auc_score(y_test, predict_pr[:,1])
```
